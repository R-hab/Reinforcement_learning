```python
# Mapping df22 columns to business_kyc schema
df_transformed = pd.DataFrame({
    "customer_id": df["_id"],
    "customer_full_name": df["fullName"],
    "customer_age": df["age"],
    "customer_gender": df["gender"],
    "customer_phone": df["phone"].astype(str),  
    "tin_number": df["tinNumber"].astype(str).fillna("N/A"),
    "house_number": df["personalAddresses[0].houseNumber"].astype(str).fillna("N/A"),
    "customer_region": df["personalAddresses[0].region"].fillna("Unknown"),
    "customer_city": df["personalAddresses[0].city"].fillna("Unknown"),
    "customer_zone_or_sub_city": df["personalAddresses[0].subcityZone"].fillna("Unknown"),
    "customer_woreda": df["personalAddresses[0].woreda"].fillna("Unknown"),
    "customer_education_level": df["educationalLevels[0].level"].fillna("Unknown"),
    "customer_marital_status": df["maritalStatuses[0]"].fillna("Unknown"),
    "business_id": None,
    "business_establishment_year": pd.to_datetime(df["businesses[0].businessRegisteredDate"], errors='coerce').dt.year.astype(str).where(pd.to_datetime(df["businesses[0].businessRegisteredDate"], errors='coerce').notna()),
    "business_sector": df["businesses[0].businessSector"].fillna("Unknown"),
    "business_starting_no_of_employees": df["businesses[0].startingNumberOfEmployees"].fillna(0).astype(int),
    "business_current_no_of_employees": df["businesses[0].numberOfEmployees"].fillna(0).astype(int),
    "business_association_type": df["businesses[0].associationType"].fillna("Unknown"),
    "business_starting_capital": df["businesses[0].startingCapital"].fillna(0).astype(float),
    "business_current_capital": df["businesses[0].currentCapital"].fillna(0).astype(float),
    "business_source_of_initial_capital": df["businesses[0].sourceOfInitialCapital"].fillna("Unknown"),
    "business_level": df["businesses[0].formOfBusiness"].fillna("Unknown"),
    "business_annual_profit": df["businesses[0].annualProfit"].fillna(0).astype(str),
    "business_registered_date": pd.to_datetime(df["businesses[0].businessRegisteredDate"], errors='coerce'),
    "business_city": df["businesses[0].businessAddresses[0].city"].fillna("Unknown"),
    "business_region": df["businesses[0].businessAddresses[0].region"].fillna("Unknown"),
    "business_subcity_zone": df["businesses[0].businessAddresses[0].subcityZone"].fillna("Unknown"),
    "business_woreda": df["businesses[0].businessAddresses[0].subcityZone"].fillna("Unknown"),
    "business_annual_income": df["businesses[0].annualSalesIncome"].fillna(0).astype(float),
    "bank": "zemzem",  # Default value
})
```

```python
dff = pd.read_csv("decision-db.scoring-history.csv")
```

```python
dff.columns
```

```python
dff.info()
```

```python
import pandas as pd

# Perform a left join on customer_id
df_merged = df_transformed.merge(dff[["customer_id", "loanProductShortName"]], 
                                 on="customer_id", 
                                 how="left")

# Fill missing values in loanProductShortName with "N/A"
df_merged["loan_product_code"] = df_merged["loanProductShortName"].fillna("N/A")

# Drop the extra loanProductShortName column if not needed
df_merged.drop(columns=["loanProductShortName"], inplace=True)
```

```python
value_counts = df_merged["loan_product_code"].value_counts()
print(value_counts)
```

```python
# Save to CSV
df_merged.to_csv("business_kyc_zemzem.csv", index=False)

print("Data transformation completed and saved as 'business_kyc.csv'.")
```

```python
df24 = pd.read_csv("decision-db.loan-processes.csv")
```

```python
df24
```

```python
# Perform a left join to bring in loan_id and status from df24
df_final = df_merged.merge(df24[["customerId", "loan_id", "status", "loanProductId"]], 
                           left_on="customer_id", 
                           right_on="customerId", 
                           how="left")

# Drop the duplicate customerId column from df24
df_final.drop(columns=["customerId"], inplace=True)
```

```python
# Save to CSV
df_final.to_csv("business_kyc_loan_process_zemzem.csv", index=False)
```
